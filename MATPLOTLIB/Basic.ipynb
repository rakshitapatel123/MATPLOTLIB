## Line Plot of Temperature Change 

```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
tempreture = [23,34,23,34,23,34,23]

plt.plot(days,tempreture)
plt.plot(days,tempreture, marker='o',linestyle ='-',color ='green',label = "Temperature")
plt.xlabel("Days of the week")
plt.ylabel("Temperature (C)")
plt.title("Weekly Temperature Change")
plt.legend()
plt.grid(color="red")
plt.show()
```

## Bar Chart of Student Marks

```python
import matplotlib.pyplot as plt
```

```python
student = ["Thulasi","Rakshita","Jay","Riya","Tanushka","Prem","Raj"]
marks = [89,78,67,98,78,56,84]
colors = ["red","blue","green","purple","orange","pink","skyblue"]
plt.barh(student,marks,color=colors,height = 0.5,)
for i in range(len(student)):
    plt.text (marks[i] + 1,i, str(marks[i]), va='center', fontsize=10, fontweight='bold')
avg_marks = sum(marks)/len(marks)
plt.axvline(x=avg_marks, color='violet', linestyle='--', linewidth=2,label=f'Avg: {avg_marks: .1f}' )
plt.xlabel("Students",fontsize = 12)
plt.ylabel("MARKS",fontsize = 12)
plt.title("Student marks",fontsize = 14,fontweight = 'bold')
plt.grid(axis ='x',linestyle ='--', alpha=0.5)
plt.legend()
plt.show()
```

## Pie Chart of Expenses

```python
import matplotlib.pyplot as plt
```

```python
expenses =[5000,2000,1500,4000,6000,1300,1400]
categories = ["Food","Travel","Shopping","Bills","Environment","Enjoying","Study"]
colors = ("pink","red","gold","silver","yellow","green","orange")
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        amount = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n{amount}"
    return my_autopct
plt.pie(expenses,labels = categories,colors=colors,autopct=make_autopct(expenses),startangle=90,shadow=True,explode=(0.05,0,0,0.1,0.1,0,0.1))
plt.title("Monthly Expenses Distribution", fontsize=14,fontweight="bold")
plt.show()
```

## Scatter Plot of Height vs Weight

```python
import matplotlib.pyplot as plt
```

```python
height = [150,155,160,165,170,175,180,185,190]
weight = [50,52,54,59,62,70,75,78,85]
scatter = plt.scatter(height,weight,c=weight,cmap = "viridis",s=[w*2 for w in weight],alpha=0.9,edgecolors="black")
cbar = plt.colorbar(scatter)
cbar.set_label("Weight (kg)")
plt.title("Scatter Plot: Height vs Weight",fontsize=14,fontweight="bold")
plt.xlabel("Height (cm)",fontsize=12)
plt.ylabel("Weight (kg)",fontsize=12)
plt.grid(True,linestyle="--",alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Data (sample)
boys_height = [160, 165, 170, 175, 180, 185]
boys_weight = [55, 60, 65, 72, 78, 85]

girls_height = [150, 155, 160, 165, 170]
girls_weight = [48, 50, 54, 58, 62]

# 2. Scatter plot (boys)
plt.scatter(boys_height, boys_weight, color="blue", marker="o", s=100, alpha=0.7, edgecolors="black", label="Boys")

# 3. Scatter plot (girls)
plt.scatter(girls_height, girls_weight, color="pink", marker="^", s=100, alpha=0.7, edgecolors="black", label="Girls")

# 4. Labels and title
plt.title("Scatter Plot: Height vs Weight (Boys vs Girls)", fontsize=14, fontweight="bold")
plt.xlabel("Height (cm)", fontsize=12)
plt.ylabel("Weight (kg)", fontsize=12)

# 5. Grid and Legend
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# 6. Show chart
plt.show()
```
